# BEER Advocate Data Extraction

```python
import markovify
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
# 400000 total
reqs = []
for i in np.arange(1, 1000, 1):
    if i % 50 == 0:
        print(i)
    reqs.append(requests.get('https://www.beeradvocate.com/beer/profile/001/' + str(i)))
```

```python
reqs
```

```python
reqs_clean = [r for r in reqs if r]
```

```python
reqs_clean
```

```python
def scrape_beer_descs(reqs):
    beer_descs = []
    for r in reqs:
        soup = BeautifulSoup(r.text, "lxml")
        beer_descs.append(get_desc_from_soup(soup))
    return beer_descs
```

```python
def get_desc_from_soup(soup):
    #return re.findall(r'\"(?P<desc>[^"]+)\"',
    info_box = soup.find('div', id='info_box')
    if not info_box:
        print('No info box')
        return None
    desc_header = info_box.find('b', string="Notes / Commercial Description:")
    if not desc_header:
        print('No Header')
        return None
    i = 0
    while i < 3:
        elem = desc_header.find_next(string=re.compile("\n\S"))
        print(elem if elem else "NA")
        if hasattr(elem, 'string') and len(elem.string) > 0:
            text = elem.string.strip('\n')
            if not text.startswith("None"):
                return text
            return None
        i += 1
    print('No String Match')
    return None
```

```python
text = scrape_beer_descs(reqs_clean)
```

```python
text = [t for t in text if t is not None]
```

```python
text_series = pd.Series(text)
```

```python
text_series.to_csv('beer_advocate_descs_1thru1000.csv')
```
